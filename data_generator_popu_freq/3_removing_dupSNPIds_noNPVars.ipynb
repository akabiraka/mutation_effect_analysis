```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
home_dir = "../"

import pandas as pd
```

```python
snp_genes_df = pd.read_csv(home_dir+"data/dbsnp/parsed/genes.tsv", sep="\t", index_col=False) # index_col=False is a must
def print_number_of_genes_attached(snps_df):
    merged_df = pd.merge(left=snps_df, right=snp_genes_df, left_on="SNP_ID", right_on="SNP_ID", how="inner")
    print("#-genes:", merged_df["GENE_NAME"].unique().shape[0])
```

```python
snps_filepath = home_dir+"data/dbsnp/parsed/snps.tsv"
snps_df = pd.read_csv(snps_filepath, sep="\t", index_col=False) # index_col=False is a must
print(snps_df.columns)
print(snps_df.shape)
```

```python
print_number_of_genes_attached(snps_df)
```

```python
print(snps_df["SNP_ID"].unique().shape[0])
print(snps_df["SNP_ID"].value_counts())
snps_df[snps_df["SNP_ID"]=="rs4317176"]#.to_csv(home_dir+"data/dbsnp/processed/rs4317176_merges_18snps.tsv", index=False, sep="\t")
```

```python
# removing data points that has the same rs-id, which are merged.
snps_dropping_merged_df = snps_df.drop_duplicates(subset="SNP_ID", keep="first")
print(snps_dropping_merged_df.shape)
print(snps_dropping_merged_df["SNP_ID"].shape)
```

```python
print_number_of_genes_attached(snps_dropping_merged_df)
```

```python
# seeing which rs-ids have no NP_VARIANTS
snps_dropping_merged_df[pd.isna(snps_dropping_merged_df["NP_VARIANTS"])]
```

```python
# removing snps where no NP_VARIANTS are present
from utils.pandas_extented_filters import filter_remove_null_nan_empty_entries
snps_removing_empty_npvar_df = filter_remove_null_nan_empty_entries(snps_dropping_merged_df, a_col_name="NP_VARIANTS") # NNE: null, nan or empty variants
print(snps_removing_empty_npvar_df.shape)
print(snps_removing_empty_npvar_df["SNP_ID"].shape[0])
```

```python
print_number_of_genes_attached(snps_removing_empty_npvar_df)
```

```python
snps_removing_empty_npvar_df.to_csv(home_dir+"data/dbsnp/processed/snps_filtered_dupSNPIds_noNPVars.tsv", index=False, sep="\t") # 7.3GB
```
