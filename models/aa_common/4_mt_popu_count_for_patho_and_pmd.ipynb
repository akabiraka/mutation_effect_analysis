```python
import os
import sys
home_dir = "../../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)

import numpy as np
import pandas as pd
from data_loader import get_merged_scores_unidirectional_df
```

```python
task = "pmd"
pmd_df = get_merged_scores_unidirectional_df(task, home_dir)
pmd_snpids_set = set(pmd_df["snp_id"].dropna().unique().tolist())
len(pmd_snpids_set)
```

```python
task = "patho"
patho_df = get_merged_scores_unidirectional_df(task, home_dir)
patho_snpids_set = set(patho_df["snp_id"].dropna().unique().tolist())
len(patho_snpids_set)
```

```python
pmd_patho_snpids_set = pmd_snpids_set.union(patho_snpids_set)
print(len(pmd_patho_snpids_set))

# all snpids are prefixed with "rs"
for snpid in pmd_patho_snpids_set:
    if "rs" not in snpid:
        print(snpid)
        break
```

```python
import gzip
with gzip.open(home_dir+"data/ALFA/freq.vcf.gz", "rt") as file_handle: #11Gs
    i=1
    data_start_flag=False
    data = []
    for line in file_handle:
        # print(line)
        if line.startswith("#CHROM"): 
            col_names = line[1:].rstrip().split('\t')# skipping the '#' symbol 
            data_start_flag=True
            continue
        elif data_start_flag==False: continue
        

        line_items = line.rstrip().split('\t')
        snpid = line_items[2]
        if snpid in pmd_patho_snpids_set:
            data.append(line_items)
        

        # if i==5: break
        if i%100000==0: print(i)
        i+=1

col_names
df = pd.DataFrame(data, columns=col_names)
df
```

```python
cols_renameing_dict = {"CHROM":"alfa_chrom_acc_version", "POS":"alfa_chrom_pos", "ID":"snp_id", "REF":"alfa_ref_allele", "ALT":"alfa_alt_allele", "SAMN10492705":"alfa_popu_count"}
popu_count_df = df[list(cols_renameing_dict.keys())].rename(columns=cols_renameing_dict)
popu_count_df
```

```python
pmd_with_popu_count_df = pd.merge(left=pmd_df, right=popu_count_df, how="left", on="snp_id")
print(pmd_with_popu_count_df.shape)
pmd_with_popu_count_df
```

```python
patho_with_popu_count_df = pd.merge(left=patho_df, right=popu_count_df, how="left", on="snp_id")
print(patho_with_popu_count_df.shape)
patho_with_popu_count_df
```

```python
patho_with_popu_count_df.to_csv(home_dir+"data/merged_predictions_unidirectional_popu_count/patho.tsv")
pmd_with_popu_count_df.to_csv(home_dir+"data/merged_predictions_unidirectional_popu_count/pmd.tsv")
# copy-paste popu_freq.tsv from "data/merged_predictions_unidirectional" to "data/merged_predictions_unidirectional_popu_count/"
```
